```python
# Her ops√¶ttes at der skal benyttes nltk biblioteket med funktionerne word_tokenize og sent_tokenize
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
sentence = "Her cat's name is Luna. Her dog's name is Max"
```

```python
sent_tokenize(sentence)
```

```python
sentence = "Hello there! How are you doing today"
word_tokenize(sentence)
```

```python
sentence_2 = "Her cat's name is Luna and her dog's name is Max."
sentence_lower = sentence_2.lower()
word_tokenize(sentence_lower)
```

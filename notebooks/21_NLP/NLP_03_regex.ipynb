```python
# Raw string - brug r foran for at undgå escape characters som \n, \t osv.
my_folder = r"C:\desktop\notes"
print(my_folder)
```

```python
# Search() - her søges efter orded "pattern" i strengen
result_search = re.search("pattern", r"string to contain pattern")
print(result_search)
```

```python
# re.sub() - erstat "pattern" med "new_pattern" i strengen
my_string = r"sara was able to help me find items quickly"
```

```python
new_string = re.sub("sara", "Sarah", my_string)
print(new_string)
```

```python
customer_reviews = ["sam was a great help to me in the store",
                   "the cashier was very rude to me, I think her name was eleanor",
                   "amazing work from sadeen!",
                   "sarah was able to help me find the items I needed quickly",
                   "lucy is such a great addition to the team",
                   "great service from sara she found me what i wanted"]
```

```python
# Find alle reviews der starter med a
a_reviews = []
pattern_to_find = r"^a"

for string in customer_reviews:
    if(re.search(pattern_to_find, string)):
        a_reviews.append(string)
```

```python
print(a_reviews)
```

```python
# Find alle reviews der slutter på y
y_reviews = []

pattern_to_find = r"y$"

for string in customer_reviews:
    if(re.search(pattern_to_find, string)):
        y_reviews.append(string)
```

```python
print(y_reviews)
```

```python
# Finde flere ord i string - 'need' 'want'
needwant_reviews = []
pattern_to_find = "r(need|want)ed"

for string in customer_reviews:
    if(re.search(pattern_to_find, string)):
        needwant_reviews.append(string)
```

```python
print(needwant_reviews)
```

```python
# Fjerne opunktum fra tekst
no_punct_reviews = []
pattern_to_find = r"[^\w\s]"
```

```python
for string in customer_reviews:
    no_punct_string = re.sub(pattern_to_find, "", string)
    no_punct_reviews.append(no_punct_string)
```

```python
print(no_punct_reviews)
```

#### Import ####

```python
from transformers import AutoTokenizer
```

###### Model med pre-trained tokenizer ######

```python
model = "bert-base-uncased"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model)
```

###### Tokenizer bruges nu på en sætning ######

```python
sentence = "I'm so exited to be learning about large language models"
```

```python
input_ids = tokenizer(sentence)
print(input_ids)
```

###### Tokenisering - opdel tekst i tokens ######

```python
tokens = tokenizer.tokenize(sentence)
```

```python
print(tokens)
```

###### Konverter tokens til input IDs ######

```python
tokens_ids = tokenizer.convert_tokens_to_ids(tokens)
```

```python
print(tokens_ids)
```

###### Konverter input IDs tilbage til tokens ######

```python
decoded_ids = tokenizer.decode(tokens_ids)
print(decoded_ids)
```

###### Brug af special tokens ######

```python
tokenizer.decode(101)
```

```python
tokenizer.decode(102)
```

#### Import ####

```python
from transformers import pipeline
```

#### Sentiment Analysis Pipeline ####

```python
sentiment_classifier = pipeline("sentiment-analysis")
```

```python
sentiment_classifier("I'm so excited to be learning about large language models")
```

#### Opret ny pipeline med NER model ####

```python
# Opret NER pipeline med en specifik model
ner = pipeline("ner", model = "dslim/bert-base-NER")
```

```python
# Brug NER pipelinen til at genkende navne og steder i en tekst
ner("Her name is Anna and she works in New York City for Morgan Stanley")
```

#### Zero-Shot Classification Pipeline ####

```python
# Opret zero-shot classification pipeline med en specifik model
zeroshot_classifier = pipeline("zero-shot-classification", model = "facebook/bart-large-mnli")
```

```python
# Sekvens der skal classificeres
sequence_to_classify = "one day I will see the world"
# Mulige labels Ai kan give teksten
candidate_labels = ['travel', 'cooking', 'dancing']
```

```python
# Brug zero-shot classification pipelinen til at klassificere sekvensen
zeroshot_classifier(sequence_to_classify, candidate_labels)
```

```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
connect_tokens = ['connecting', 'connected', 'connection', 'connections', 'connects']
```

```python
# Stemming af tokens
for token in connect_tokens:
    print(token, ": ", ps.stem(token))
```

```python
# Stemming af tokens der er forskellige
likes_tokens = ['likes', 'better', 'worse']
```

```python
for token in likes_tokens:
    print(token, ": ", ps.stem(token))
```
